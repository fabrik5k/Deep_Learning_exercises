Sobre o modelo
Usando o dataset load iris, fizemos a seguinte rede neural
Uma Rede Neural DENSA cujo objetivo é classificar a que classe pertence a iris de cada linha do dataset x
Foi utilizada a função de ativação ReLu para as camadas ocultas, e para a camada de saída foi utilizada a softmax
A rede neural possui 4 neuronios na camada de entrada pelo fato de possuir 4 features (ou variaveis, escolha o nome que preferir)
Possui uma camada oculta com 10 neuronios que recebem os valores dos neuronios de entrada
E a camada de saída com 3 neuronios respectivos a quantidade de classes que temos na saída y do dataset
Como otimizador foi utilizado o gradiente descendente estocástico

Compilação do modelo
A métrica usada para validação foi a métrica da acuracia, pois se encaixa perfeitamente em nosso modelo, visto que temos uma igualdade de frequência nos valores de y 
A função de erro(ou loss, como preferir) foi definida como categorical_crossentropy
O otimizador como o Gradiente descendente estocàstico

Treinamento do modelo
O modelo foi treinado com 1000 epochs ou seja 1000 vezes em que nosso modelo foi treinado com a base de dados
Com o tamanho dos dados em que será aplicado o gradiente descendente estocástico sendo igual ao tamanho real do dataframe, 105 no total, então teremos apenas uma iteração por epoch logo os pesos e os bias foram alterados um total de 1000 vezes dentro da rede neural
Utilizando os dados de teste como validação para verificar a presença de overfiting dado que o dataset não é tão grande o que nos permite fazer isso

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
x = pd.DataFrame(iris.data, columns=[iris.feature_names])
y = pd.Series(iris.target)
```

```python
x.head()
```

```python
y.value_counts()
```

```python
from keras.utils import to_categorical

y_convertido = to_categorical(y)
y_convertido
```

```python
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y_convertido, test_size=0.3)
```

```python
from keras.layers import Activation
from keras.optimizers import SGD
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
modelo = Sequential()

modelo.add(Dense(10, kernel_initializer='normal', input_dim=4, activation='relu'))

modelo.add(Dense(3, kernel_initializer='normal', activation='softmax'))
```

```python
otimizador = SGD()
```

```python
modelo.compile(loss='categorical_crossentropy', optimizer=otimizador, metrics=['acc']) #"acc" é a métrica da acurácia
```

```python
modelo.fit(x_treino, y_treino, epochs=1000, batch_size=105, validation_data=(x_teste, y_teste), verbose=1)
#epochs é o numero de iterações, a quantidade de vezes que será mudado o valor das variáveis (pesos e bias)
#batch_size é o tamanho do conjunto de dados que eu vou usar pra treinar meu modelo (O tamanho do gradiente descendente estocastico)
#Ele vai imprimir varias mensagens na tela enquanto a rede neural é treinada
```
